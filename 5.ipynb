–ü—Ä–∏–Ω—Ü–∏–ø –≤—ã–±–æ—Ä–∞ –ø–æ–∑–∏—Ü–∏–∏ –≤ —Å—Ç—Ä–æ–∫–µ –¥–ª—è –ø–µ—á–∞—Ç–∏ ‚Äú–∑–≤—ë–∑–¥–æ—á–∫–∏‚Äù –¥–æ–ª–∂–µ–Ω –±—ã—Ç—å —Å–ª–µ–¥—É—é—â–∏–º:
–¥–∏–∞–ø–∞–∑–æ–Ω –∑–Ω–∞—á–µ–Ω–∏–π —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–∞ –∑–∞–¥–∞–Ω–Ω–æ–º –æ—Ç—Ä–µ–∑–∫–µ –¥–µ–ª–∏—Ç—Å—è –Ω–∞ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–Ω–∞–∫–æ–º–µ—Å—Ç

–¥–∏–∞–ø–∞–∑–æ–Ω–∞ –∑–Ω–∞—á–µ–Ω–∏–π –¥–ª—è –∫–∞–∂–¥–æ–≥–æ –∑–Ω–∞–∫–æ–º–µ—Å—Ç–∞. –ù–∞–ø—Ä–∏–º–µ—Ä, –¥–ª—è —Ñ—É–Ω–∫—Ü–∏–∏ y=x –Ω–∞ –æ—Ç—Ä–µ–∑–∫–µ
–æ—Ç 0 –¥–æ 100 –ø—Ä–∏ —à–∏—Ä–∏–Ω–µ –≥—Ä–∞—Ñ–∏–∫–∞ 80 –æ–¥–Ω–æ–º—É –∑–Ω–∞–∫–æ–º–µ—Å—Ç—É –±—É–¥–µ—Ç —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–æ–≤–∞—Ç—å
–¥–∏–∞–ø–∞–∑–æ–Ω 100/80=1.25. –¢–æ–≥–¥–∞ ‚Äú–∑–≤—ë–∑–¥–æ—á–∫–∞‚Äù –≤ –ø–µ—Ä–≤–æ–π –ø–æ–∑–∏—Ü–∏–∏ –¥–æ–ª–∂–Ω–∞ –≤—ã–≤–æ–¥–∏—Ç—å—Å—è, –µ—Å–ª–∏
–∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –ø–æ–ø–∞–¥–∞–µ—Ç –≤ –ø–æ–ª—É–∏–Ω—Ç–µ—Ä–≤–∞–ª [0, 1,25), –≤–æ –≤—Ç–æ—Ä–æ–π –ø–æ–∑–∏—Ü–∏–∏ –¥–ª—è
–ø–æ–ª—É–∏–Ω—Ç–µ—Ä–≤–∞–ª–∞ [1,25, 2,5), –∏ —Ç. –¥.
–ù–∞–ø–∏—Å–∞—Ç—å –ø—Ä–æ–≥—Ä–∞–º–º—É —Ç—Ä–µ–±—É–µ—Ç—Å—è —Å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏–µ–º —É–∫–∞–∑–∞–Ω–Ω–æ–≥–æ –ø–æ –≤–∞—Ä–∏–∞–Ω—Ç—É —Ü–∏–∫–ª–∞. –¢–∞–∫–∂–µ
–Ω–µ–æ–±—Ö–æ–¥–∏–º–æ –≤—ã–ø–æ–ª–Ω–∏—Ç—å –¥–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ.

–§—É–Ω–∫—Ü–∏–∏:
$$ùëÉ = ùë†‚Ñé(ùë•) + ùëêùëúùë†2ùë•,   –≥–¥–µ ùë†‚Ñé(ùë•) = \frac{1}{2}\ * (ùëí^ùë• ‚àí ùëí^{‚àíùë•})$$  
$$ùëÑ = ùëí^{‚àíùë•+1} + 1$$
–ê—Ä–≥—É–º–µ–Ω—Ç:
|–ù–∞—á–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ| –®–∞–≥ |–ö–æ–Ω–µ—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ|
|---|---|---|
|$$ùë•_0 =‚àí 1$$|$$‚Ñé = 0. 05$$|$$ùë•_ùëõ = 1$$|  

–û–ø–µ—Ä–∞—Ç–æ—Ä: while
–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ–µ –∑–∞–¥–∞–Ω–∏–µ:
1) –æ–ø—Ä–µ–¥–µ–ª–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏—è $P_{min}$ –∏ $x_{min}$, –ø—Ä–∏ –∫–æ—Ç–æ—Ä–æ–º –æ–Ω–æ –¥–æ—Å—Ç–∏–≥–∞–µ—Ç—Å—è.

```python
from math import (
    pow,
    sinh,
    cos,
    e
)

def shFunc(x: float) -> float: # —Ä–∞—Å—á–µ—Ç –∑–Ω–∞—á–µ–Ω–∏—è 1 —Ñ—É–Ω–∫—Ü–∏–∏
    y = sinh(x) + cos(2*x)
    # y = x
    return y

def eFunc(x: float) -> float: # —Ä–∞—Å—á–µ—Ç –∑–Ω–∞—á–µ–Ω–∏–µ 2 —Ñ—É–Ω–∫—Ü–∏–∏
    y = pow(e, -x + 1) + 1
    return y

def visualizeFunc(funcValues: list[tuple[float, float]]) -> None: # —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏–∏
    global numberOfSerifs, SIZEY
    
    # –º–∞–∫—Å–∏–º—É–º –º–∏–Ω–∏–º—É–º –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é —Ñ—É–Ω–∫—Ü–∏–∏
    Xmin, minValues = min(funcValues, key = lambda x: x[1])
    Xmax, maxValues = max(funcValues, key = lambda x: x[1])
    print(f'–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏: {minValues:.7g}, –ø—Ä–∏ X: {Xmin:.7g}',
          f'–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏: {maxValues:.7g}, –ø—Ä–∏ X: {Xmax:.7g}',
          '-' * (SIZEY + 14),
          sep='\n')
    
    oneSerifSize = (maxValues - minValues) / (numberOfSerifs - 1) # —à–∞–≥ –¥–ª –º–µ—Ç–æ–∫
    onePixelSize = (maxValues - minValues) / SIZEY # —Ä–∞–∑–º–µ—Ä —Å–∫–æ–ª—å–∫–æ –ø–æ–º–µ—â–∞–µ—Ç—Å—è –≤ 1 —Å–∏–º–≤–æ–ª–µ
    
    s = ' ' * 6
    for i in range(numberOfSerifs):
        # –≤—ã–≤–æ–¥ –º–µ—Ç–æ–∫
        currentSerif = minValues + i*oneSerifSize
        serifPosition = int((i * oneSerifSize) // onePixelSize)
        s += f'{' '*(serifPosition - len(s) + 5)}{(currentSerif):^6.4g}'
    print(s)
    
    # –ø—Ä–æ—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ–∑–∏—Ü–∏—è 0
    zeroPosition = 0
    if minValues < 0 < maxValues:
        zeroPosition = int(-minValues // onePixelSize)
    
    for x, y in funcValues:
        # –ø—Ä–æ—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ–∑–∏—Ü–∏—è –Ω—É–∂–Ω–æ–π —Ç–æ—á–∫–∏
        yPosition = int((y - minValues) // onePixelSize)
        # –≤—ã–≤–æ–¥ —Å —É—á–µ—Ç–æ–º
        if zeroPosition:
            if yPosition < zeroPosition:
                print(f'{x:>7.2f}|{' ' * yPosition}*{' ' * (zeroPosition - yPosition - 1)}|')
            elif yPosition > zeroPosition:
                print(f'{x:>7.2f}|{' ' * zeroPosition}|{' ' * (yPosition - zeroPosition - 1)}*')
            else: 
                print(f'{x:>7.2f}|{' ' * yPosition}*')
        else:
            print(f'{x:>7.2f}|{' ' * yPosition}*')
```

```python
EPS = 1e-8
SIZEY = 120

# –≤—Å–µ –≤–≤–æ–¥—ã
beginValueX = float(input('–ù–∞—á–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ x: '))
endValueX = float(input('–ö–æ–Ω–µ—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ x: '))
step = float(input('–®–∞–≥ –ø–æ x: '))
numberOfSerifs = int(input('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–∞—Å–µ—á–µ–∫: '))
```

```python
func1Values = []
func2Values = []

# –æ–≥–ª–∞–≤–ª–µ–Ω–∏–µ —Ç–∞–±–ª–∏—Ü—ã
print('-'*44, '\n',
      f"|{'x':^10}|{'y1':^15}|{'y2':^15}|", '\n',
     '-'*44, sep='')

curentX = beginValueX

while curentX - endValueX <= EPS: # —Å—á–µ—Ç—á–∏–∫ –ø–æ–∫–∞ –Ω–∞—Å—Ç–æ—è—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ <= –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ
    y1 = shFunc(curentX)
    y2 = eFunc(curentX)
    
    # –ø—Ä–∏—Å–≤–∞–∏–≤–∞–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π –≤ –º–∞—Å—Å–∏–≤
    func1Values.append((curentX, y1))
    func2Values.append((curentX, y2))
    
    # –≤—ã–≤–æ–¥ –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Ç–∞–±–ª–∏—Ü—É
    print(f'|{curentX:^10.2f}|{y1:^15.7g}|{y2:^15.7g}|') 
    
    curentX += step # —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ + —à–∞–≥

print('-'*44)
```

```python
funcValues = func1Values

# –º–∞–∫—Å–∏–º—É–º –º–∏–Ω–∏–º—É–º –ø–æ –∑–Ω–∞—á–µ–Ω–∏—é —Ñ—É–Ω–∫—Ü–∏–∏
Xmin, minValues = min(funcValues, key = lambda x: x[1])
Xmax, maxValues = max(funcValues, key = lambda x: x[1])
print(f'–ú–∏–Ω–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏: {minValues:.7g}, –ø—Ä–∏ X: {Xmin:.7g}',
        f'–ú–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏: {maxValues:.7g}, –ø—Ä–∏ X: {Xmax:.7g}',
        '-' * (SIZEY + 14),
        sep='\n')

oneSerifSize = (maxValues - minValues) / (numberOfSerifs - 1) # —à–∞–≥ –¥–ª –º–µ—Ç–æ–∫
onePixelSize = (maxValues - minValues) / SIZEY # —Ä–∞–∑–º–µ—Ä —Å–∫–æ–ª—å–∫–æ –ø–æ–º–µ—â–∞–µ—Ç—Å—è –≤ 1 —Å–∏–º–≤–æ–ª–µ

s = ' ' * 6
for i in range(numberOfSerifs):
    # –≤—ã–≤–æ–¥ –º–µ—Ç–æ–∫
    currentSerif = minValues + i*oneSerifSize
    serifPosition = int((i * oneSerifSize) // onePixelSize)
    s += f'{' '*(serifPosition - len(s) + 5)}{(currentSerif):^6.4g}'
print(s)

# –ø—Ä–æ—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ–∑–∏—Ü–∏—è 0
zeroPosition = 0
if minValues < 0 < maxValues:
    zeroPosition = int(-minValues // onePixelSize)

for x, y in funcValues:
    # –ø—Ä–æ—Å—á–∏—Ç—ã–≤–∞–µ—Ç—Å—è –ø–æ–∑–∏—Ü–∏—è –Ω—É–∂–Ω–æ–π —Ç–æ—á–∫–∏
    yPosition = int((y - minValues) // onePixelSize)
    # –≤—ã–≤–æ–¥ —Å —É—á–µ—Ç–æ–º
    if zeroPosition:
        if yPosition < zeroPosition:
            print(f'{x:>7.2f}|{' ' * yPosition}*{' ' * (zeroPosition - yPosition - 1)}|')
        elif yPosition > zeroPosition:
            print(f'{x:>7.2f}|{' ' * zeroPosition}|{' ' * (yPosition - zeroPosition - 1)}*')
        else: 
            print(f'{x:>7.2f}|{' ' * yPosition}*')
    else:
        print(f'{x:>7.2f}|{' ' * yPosition}*')
```
