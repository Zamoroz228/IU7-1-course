–≤–∞—Ä–∏–∞–Ω—Ç—É), —Ç–æ—á–Ω–æ—Å—Ç—å, –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç–µ—Ä–∞—Ü–∏–π –∏ —à–∞–≥ –ø–µ—á–∞—Ç–∏. –ù–µ–æ–±—Ö–æ–¥–∏–º–æ
–≤—ã–≤–µ—Å—Ç–∏ —Ç–∞–±–ª–∏—Ü—É –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω—ã—Ö –≤—ã—á–∏—Å–ª–µ–Ω–∏–π —Å –∑–∞–¥–∞–Ω–Ω—ã–º —à–∞–≥–æ–º (–Ω–æ–º–µ—Ä –∏—Ç–µ—Ä–∞—Ü–∏–∏,
–∑–Ω–∞—á–µ–Ω–∏–µ —Ç–µ–∫—É—â–µ–≥–æ —á–ª–µ–Ω–∞, –ø—Ä–æ–º–µ–∂—É—Ç–æ—á–Ω–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Å—É–º–º—ã) –∏ —Ä–µ–∑—É–ª—å—Ç–∞—Ç - –≤—ã—á–∏—Å–ª–µ–Ω–Ω–æ–µ
–∑–Ω–∞—á–µ–Ω–∏–µ —Å—É–º–º—ã —Ä—è–¥–∞ –ª–∏–±–æ —Å–æ–æ–±—â–µ–Ω–∏–µ –æ —Ç–æ–º, —á—Ç–æ –∑–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –∏—Ç–µ—Ä–∞—Ü–∏–π
–Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–π —Ç–æ—á–Ω–æ—Å—Ç–∏ –¥–æ—Å—Ç–∏—á—å –Ω–µ —É–¥–∞–ª–æ—Å—å.
$$ùë¶ = \frac{1}{2^2} + \frac{1}{5^2} + \frac{1}{8^2} + ‚Ä¶ + \frac{1}{(3n-1)^2} + ‚Ä¶ $$

```python
EPS = float(input('–í–≤–µ–¥–∏—Ç–µ —Ç–æ—á–Ω–æ—Å—Ç—å —Ä–∞—Å—á–µ—Ç–æ–≤: '))
MAXITT = int(input('–í–≤–µ–¥–∏—Ç–µ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç—Ç–µ—Ä–∞—Ü–∏–π: '))
STEPOUTPUT = int(input('–í–≤–µ–¥–∏—Ç–µ —à–∞–≥ –≤—ã–≤–æ–¥–∞: '))
```

```python
sumSequence = 0

# –∑–∞–≥–æ–ª–æ–≤–æ–∫ —Ç–∞–±–ª–∏—Ü—ã
print('-'*44, '\n',
      f"|{'‚Ññ –∏—Ç–µ—Ä–∞—Ü–∏–∏':^10}|{'t':^15}|{'s':^15}|", '\n',
     '-'*44, sep='')

for i in range(1, MAXITT + 1):
    
    currentValue = 1 / (3*i - 1)**2 # —Ç–µ–∫—É—â–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
    sumSequence += currentValue
    
    if (i - 1) % STEPOUTPUT == 0: # –ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ —à–∞–≥ –ø–µ—á–∞—Ç–∏
        print(f'|{i:^10}|{currentValue:^15.7g}|{sumSequence:^15.7g}|')
        
    if currentValue <= EPS: # –ø—Ä–æ–≤–µ—Ä–∫–∞ –Ω–∞ –Ω—É–∂–Ω—É—é —Ç–æ—á–Ω–æ—Å—Ç—å
        print('-'*44, '\n',
              f'–°—É–º–º–∞ –±–µ—Å–∫–æ–Ω–µ—á–Ω–æ–≥–æ —Ä—è–¥–∞ - {sumSequence}, –≤—ã—á–∏—Å–ª–µ–Ω–∞ –∑–∞ {i} –∏—Ç–µ—Ä–∞—Ü–∏–∏.',
              sep='')
        break
else: # –µ—Å–ª–∏ –Ω–µ –¥–æ—Å—Ç–∏–≥–Ω—É—Ç–æ –∑–∞ –Ω—É–∂–Ω–æ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∏—Ç—Ç–µ—Ä–∞—Ü–∏–π
    print('-'*44, '\n'
          '–ó–∞ —É–∫–∞–∑–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –∏—Ç–µ—Ä–∞—Ü–∏–π –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ–π —Ç–æ—á–Ω–æ—Å—Ç–∏ –¥–æ—Å—Ç–∏—á—å –Ω–µ —É–¥–∞–ª–æ—Å—å.',
          sep='')
```
