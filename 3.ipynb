#### Задание 1
Написать программу, которая по введенному значению аргумента x
определит значение функции у.   

![Image](1.png)

```python
from math import log10 # импорт функции из библиотеки math

def function1WithLogarithm(x): # функция для просчета первой функции
    if x <= -5: # проверка на область определения
        return None
    y = -log10(x + 5)
    return y

def function2ComplexSqrt(x): # функция для просчета второй функции 
    if x < -4 or x > 0: # проверка на область определения
        return None
    y = (4 - (x + 2)**2)**0.5
    return y

def function3BasicSqrt(x): # функция для просчета третий функции
    if x < 0: # проверка на область определения
        return None
    y = x**0.5
    return y

def function4Lighest(x): # функция для просчета четвертой функции
    y = -x + 6
    return y

x = float(input('Введите x: ')) # ввод точки х с клавиатуры

'''
условные операторы для распределения функций по их кусочкам 
области определения
'''
if x <= -5: 
    print('Функция не имеет решений')
    exit()
elif x < -4:
    y = function1WithLogarithm(x)
elif x < 0:
    y = function2ComplexSqrt(x)
elif x <= 4:
    y = function3BasicSqrt(x)
else:
    y = function4Lighest(x)

print(f'Функция имеет значение {y:.7g} в данной точке') # вывод значения функции
```

Написать программу, которая по введенным значениям координат точки (x,
y) определит, принадлежит ли точка заданной замкнутой области, включая границы. 

![Image](2.png)  
Заменим первую и вторую функции на y = -2|x| + 4 ({-6 <= x <= 6})

```python
def func1(x): # функция для графика y = -2|x| + 4
    y = -2 * abs(x) + 4
    return y

def func2(x): # функция для графика y = -1/4 * x^2 + 1
    y = -1/4 * x**2 +1
    return y

x, y = map(float, input('Введите координаты точки через пробел').split()) # ввод данных с клавиатуры

textPointLies = "Точка лежит внутри или на линии пересечения" # текст если точка лежит внутри
textPointDoesntLies = "Точка не лежит внутри или на линии пересечения" # обратный текст

if -2 <= x <= 2: # проверка на первый верхний промежуток 
    if func1(x) >= y and func2(x) <= y: # 1 функция выше, 2 ниже 
        print(textPointLies)
    else:
        print(textPointDoesntLies)
elif -6 <= x <= 6: # проверка на два боковых промежутка
    if func1(x) <= y and func2(x) >= y: # 1 функция ниже, 2 выше
        print(textPointLies)
    else:
        print(textPointDoesntLies)
else:
    print(textPointDoesntLies)
```
